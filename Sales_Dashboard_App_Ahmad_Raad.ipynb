```python
import pandas as pd
import dash
from dash import dcc, html
import plotly.express as px
```

```python
# Load the uploaded dataset
sales_data = pd.read_csv('sales_data.csv')

# Display the first few rows of the dataset
print (sales_data.head())

# Calculate basic statistics
print("\nTotal Sales:", sales_data['sales amount'].sum())
print("\nSales by Store Location:")
print(sales_data.groupby('store location')['sales amount'].sum())

# Aggregate sales by store location
sales_by_location = sales_data.groupby('store location')['sales amount'].sum().reset_index()

# Initialize the Dash app
app = dash.Dash(__name__)

# Create a bar chart using Plotly Express
fig = px.bar(sales_by_location, x='store location', y='sales amount', title='Sales by Store Location')

# Define the layout of the app
app.layout = html.Div(children=[
    html.H1(children='Sales Dashboard'),
    dcc.Graph(figure=fig)
])

# Run the Dash app
if __name__ == '__main__':
    app.run_server(debug=True)
```
